# The geometry of large neural population manifold

# 1 - Manifold dimensionality

* “dimensionality is a complexity measure of the population activity and may be suggestive of underlying circuit mechanisms.” ([Cunningham and Yu, 2014, p. 1506](zotero://select/groups/5234466/items/4JNUMGKU)) ([pdf](zotero://open-pdf/groups/5234466/items/44MYHH9M?page=7&annotation=FHL5EXH2))

* **Why manifold dimensionality matters**:
  * evaluate how complete the visualized manifold is in a 2D or 3D plots (is maximal if most of the variance in the data is explained by the first 2 or 3 principal components). If its dimensionality is higher than 3, we know that some amount of information is lost when the data's dimensionality is reduced  in 2D or 3D for visualization.
  * estimate the computational complexity of decoding
  * predict decoding accuracy: high-dimensional datasets
  * Low-dimensional latent signals can facilitate the manipulation or the extraction of signals for brain-computer interfaces, a rehabilitative technology that converts neural signals into control commands to restore movement to paralyzed patients (see (2)).
  * higher dimensionality may provide an advantage for downstream neurons reading out information from the recorded population ([Cunningham and Yu, 2014, p. 1506](zotero://select/groups/5234466/items/4JNUMGKU)) ([pdf](zotero://open-pdf/groups/5234466/items/44MYHH9M?page=7&annotation=FHL5EXH2), see (2))

* strong tuning
  * Extreme case: "labelled line" code: tuning function where each stimulus elicit a response from a single neuron
    * same dimensionality as the number of neurons N, it can move freely in all N directions of the neural state space
    * this yields a "splitter code"

* Weak tunings
  * approximates a circle manifold, which can be fully described with only two dimensions (two principal components). The manifold is constrained to a linear subspace (the `intrinsinc_manifold`) of the N neuron state space.
