## **High dimensionality**


### Advantages

* higher dimensionality may provide an advantage for downstream neurons reading out information from the recorded population [Cunningham, 2014, Rigotti, 2013].


* Nonlinear mixed selectivity is a factor of high-dimensional neural code, highly valuable to solve complex cognitive tasks [Rigotti, 2013].


* Complex cognitive tasks such as `contextual tasks` (e.g., rule-based tasks), with high dimension [Rigotti, 2013]


* The responses of a population of nonlinear mixed selective neurons can be classified with different hyperplanes (the position of which changes with each task rule), but not the responses of a population of pure linear mixed neurons [Rigotti, 2013].


* High dim representations allow simple readouts such as linear classifiers to implement a large set of input-output classifications [Rigotti, 2013].

* Model circuits that rely on such high-dimensional representations can generate very rich dynamics [Rigotti, 2013]

* The same property is exploited in contemporary machine learning techniques such as support vector machine [Rigotti, 2013].

### Limitations

* nonlinear mixed selective neurons are more sensitive to task-irrelevant sources of variability (produce more errors) than pure and linear mixed selective neurons [Rigotti, 2013].

## **The factors that determine dimensionality**

### Tuning strength

#### Very strong tunings

  * The dimensionality is N, the total number of neurons.
  
  * The **Labelled line code** is the extreme case: each stimulus elicits a response from a single neuron which produces high dimensionality.
  
  * **Sparse units** each selective to a unique property value of each task variable can produce high-dimensionality
  ("Grandmother cells", [Rigotti 2013])
  
  * In these cases the manifold can move freely in all N directions of the neural state space.
    

#### Weak tunings

  * They produce a **ring** (circular manifold), which can be fully described with only two dimensions (two principal components).
  
  * The manifold is constrained to a linear subspace (the `intrinsinc_manifold`) of the N neural state space.
